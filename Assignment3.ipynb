```
import pandas as pd
import numpy as np
```

```
data=pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/Loan_Data/loan_train.csv')
data.head()
```

```
data.drop(columns=['Unnamed: 0'	,'Loan_ID'], axis=1, inplace=True)
```

```
data.head()
```

```
data.shape
```

```
from pycaret.classification import *
```

```
exp=setup(data=data, target='Loan_Status', feature_selection=True)
```

```
models()
```

```
top8=compare_models(n_select=8, sort='F1')
```

```
tuned_top8=[tune_model(i) for i in top8]
```

```
blender=blend_models(tuned_top8)
```

```
stacker=stack_models(tuned_top8)
```

```
test_data = pd.read_csv('https://raw.githubusercontent.com/dphi-official/Datasets/master/Loan_Data/loan_test.csv')
```

```
new_test_data=test_data.iloc[0:1,:]
new_test_data
```

```
y_pred_final=predict_model(blender, data=new_test_data)
```

```
y_pred_final.head()
```

```
y_pred_final['Credit_History'].min(), y_pred_final['Credit_History'].max()
```

```
predictions=y_pred_final['Label']

y_pred_final=pd.Series(predictions, index=None, name='prediction')
y_pred_final.to_csv('results.csv', index=None)
```
